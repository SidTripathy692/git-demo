```python
class Atm:
    def __init__(self):
        self.balance = 0
        self.pin = ""
        self.menu()

    def menu(self):
        user_input = input("""
        1. Create PIN
        2. Change PIN
        3. Deposit
        4. Withdraw
        5. Balance Inquiry
        6. Exit
        """)

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.deposit()
        elif user_input == '4':
            self.withdraw()
        elif user_input == '5':
            self.balance_inquiry()
        elif user_input == '6':
            print("Thanks for banking with us!")
            exit()
        else:
            print("Invalid Input")

    def create_pin(self):
        self.pin = input("Enter the PIN: ")
        self.balance = int(input("Enter the balance: "))
        print("PIN created successfully.")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter the old PIN: ")
        if old_pin == self.pin:
            new_pin = input("Enter the new PIN: ")
            self.pin = new_pin
            print("PIN changed successfully.")
            self.menu()
        else:
            print("Incorrect PIN")
            self.menu()

    def deposit(self):
        amount = int(input("Enter amount to be deposited: "))
        self.balance += amount
        print("Rs.", amount, "successfully deposited.")
        print("Available balance:", self.balance)
        self.menu()

    def withdraw(self):
        amount = int(input("Enter amount to be withdrawn: "))
        user_pin = input("Enter your PIN: ")
        if user_pin == self.pin:
            if amount <= self.balance:
                self.balance -= amount
                print("Rs.", amount, "successfully debited.")
                print("Available balance:", self.balance)
            else:
                print("Insufficient balance.")
        else:
            print("Incorrect PIN")
        self.menu()

    def balance_inquiry(self):
        user_pin = input("Enter your PIN: ")
        if user_pin == self.pin:
            print("Available balance:", self.balance)
            self.menu()
        else:
            print("Incorrect PIN")
            self.menu()
obj = Atm()
```
